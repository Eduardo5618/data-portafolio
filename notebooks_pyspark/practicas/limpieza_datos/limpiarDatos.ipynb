```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
from pyspark.sql.functions import col,desc


spark = SparkSession.builder.appName('prueba').getOrCreate()

sc = spark.sparkContext

```

```python
df = spark.read.parquet('../../data/datos.parquet')
```

```python
df.count()
```

```python
# EL TRUNCATE FALSE hace que me muestre todo y no se corte
df.show(5,truncate=False)
```

```python
df.select('video_id').collect()
```

```python
#ELIMINADO LAS FILAS QUE ALMENOS TIENEN UN VALOR NULO
df.na.drop().count()
```

```python
# Esto borrara los valores donde LA COLUMNA VIEWS habia un valor nulo
df.na.drop(subset=['views']).count()
```

```python
df.na.drop(subset=['views','dislikes']).count()
```

```python
df.orderBy(col('views')).select(col('views'),col('likes'),col('dislikes')).show()
```

```python
# Aqui las RELLENAMOS CON 0
df.fillna(0).orderBy(col('views')).select(col('views'),col('likes'),col('dislikes')).show()
```

```python
# Tambien podemos seleccionar una COLUMNA
df.fillna(0,subset=['likes']).orderBy(col('views')).select(col('views'),col('likes'),col('dislikes')).show()
```
